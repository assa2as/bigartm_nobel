```python
plus, topics, min_df = 'booksX2', 300, 1
path, file_ = 'models/books/','books'

lst_topics = dict()
with open(path+plus+'-'+str(topics)+'_'+str(min_df)+'.topics') as f:
    for line in f:
        vals = line.strip().split(' ')
        lst_topics[int(vals[0])] = ' '.join(vals[1:])
def getTem(tem):
    return lst_topics[tem]
```

```python
n = 0
tems0 = dict()
with open(path+plus+'-'+str(topics)+'_'+str(min_df)+'.tems') as f:
    for line in f:
        vals = line.strip().split('| ')
        name = vals[0]
        tems = sorted([x.split('^') for x in vals[1].split(' ')],key=lambda f: (-float(f[1]),int(f[0])))
        for tem,weight in tems:
            x, y = int(tem), float(weight)
            z = tems0.get(x,[0,0.,0.,dict()])
            z[0] += 1
            z[1] += y
            z[2] = max(z[2],y)
            z[3][name] = y 
            tems0[x] = z
        print(name.replace('_',' '))
        for tem,weight in tems[:3]:
            print(' ',getTem(int(tem)),'('+str(weight)+')')
        n += 1
```

```python
for tem,stat in sorted(tems0.items(),key=lambda f: (-f[1][2],-f[1][0])):
    print(getTem(tem),'['+str(stat[0])+' книг]')
    for name,w in sorted(stat[3].items(),key=lambda f: -f[1])[:10]:
        print('  ',name.replace('_',' '),'('+str(w)+')')
```
